```python
import requests
import bs4
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/MasterChef_Australia_(series_13)"
requests.get(url)
```

```python
# requests.get(url,{}).text
```

```python
web_page= bs4.BeautifulSoup(requests.get(url,{}).text)
```

```python
tables = web_page.find_all(name="tbody")[1]
tables = tables('tr')
tables[1]
```

```python
columns = ["Contestant","Age","State/Terr","Occupation", "Status"]
```

```python
temp_list=[]
for table in tables:
    data= table.find_all(name="td")
    temp_list.append([x.text[:-1] for x in data])
# del temp_list[8:10]
print(temp_list)
df = pd.DataFrame(data=temp_list,columns=columns)
df
```
